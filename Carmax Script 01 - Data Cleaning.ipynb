```python
# import libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# create path variable for imported csv file
carmax_data_import_path = "Carmax - Main (Original).csv"

# read in csv file as carmax_data_original using import path variable
carmax_data_original = pd.read_csv(carmax_data_import_path) 

# print result for carmax_data_original
carmax_data_original  
```

```python
# create backup of carmax_data_original as carmax_modified
carmax_modified = carmax_data_original.copy() 
```

```python
# check for presence of NaN values
if len(carmax_modified.dropna()) == len(carmax_modified):
    print("No missing values")
else:
    print("Missing values")
```

```python
# print the sum of NaN values in each column
carmax_modified.isnull().sum()
```

```python
# plot the sum of NaN values in each column
carmax_modified.isnull().sum().plot(kind="bar")
plt.show()
```

```python
# print data types for values each column
carmax_modified.dtypes 
```

```python
# print columns to list
carmax_modified.columns.tolist() 
```

```python
# rename columns
carmax_modified.rename(columns = {'purchase_make':'vehicle_make', 'purchase_model':'vehicle_model', 'purchase_vehicle_year':'vehicle_year', 'purchase_price':'purchase_price_bins', 'purchase_price_mid_range':'purchase_price', 'vehicle_financing':'financing', 'customer_age':'customer_age_bins', 'customer_age_mid_range':'customer_age', 'customer_income':'customer_income_bins', 'customer_income_mid_range':'customer_income', 'customer_previous_purchase':'previous_purchase', 'customer_distance_to_dealer':'distance_to_dealer', 'vehicle_warranty_used' :'warranty'}, inplace = True) 
carmax_modified 
```

```python
# print data types for each column
carmax_modified.dtypes
```

```python
# create list of columns to keep
columns_to_keep = ['insert_num', 'vehicle_make', 'vehicle_model', 'vehicle_year', 
'purchase_price', 'trade_in', 'financing', 'customer_age', 'customer_income', 'customer_gender', 
'previous_purchase', 'distance_to_dealer', 'post_purchase_satisfaction', 'warranty', 'subsequent_purchases']
```

```python
# convert object datatypes to float64 datatypes
carmax_modified["distance_to_dealer"] = pd.to_numeric(carmax_modified.distance_to_dealer, errors='coerce')
carmax_modified["post_purchase_satisfaction"] = pd.to_numeric(carmax_modified.post_purchase_satisfaction, errors='coerce') 

# convert int64 datatypes to boolean datatypes
carmax_modified["trade_in"] = carmax_modified["trade_in"].astype('bool')
carmax_modified["financing"] = carmax_modified["financing"].astype('bool')
carmax_modified["previous_purchase"] = carmax_modified["previous_purchase"].astype('bool') 
carmax_modified["warranty"] = carmax_modified["warranty"].astype('bool')
```

```python
# print converted datatypes 
carmax_modified.dtypes
```

```python
# create filtered data frame with select columns as carmax_filtered
carmax_filtered = carmax_modified[columns_to_keep]
carmax_filtered 
```

```python
# print the sum of NaN values in each column
carmax_filtered.isnull().sum()
```

```python
# plot the sum of NaN values in each column
carmax_filtered.isna().sum().plot(kind="bar")
plt.show()
```

```python
# replace NaN values in distance_to_dealer column
carmax_filtered['distance_to_dealer'].fillna(0, inplace=True) 
```

```python
# replace NaN values in post_purchase_satisfaction column
carmax_filtered['post_purchase_satisfaction'].fillna(0, inplace=True) 
```

```python
# check for presence of NaN values after replacement
if len(carmax_filtered.dropna()) == len(carmax_filtered):
    print("No missing values")
else:
    print("Missing values")
```

```python
# print result of carmax_filtered
carmax_filtered
```

```python
# create path variable for output csv file
carmax_output_path = "Carmax - Main (Modified).csv"
# convert carmax_filtered to csv file using output path variable
carmax_output_file = carmax_filtered.to_csv(carmax_output_path)
```
