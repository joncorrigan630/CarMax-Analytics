```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
from matplotlib import pyplot as plt

import_file_path = "Carmax - Main (Modified).csv"
data = pd.read_csv(import_file_path, index_col=0)
df = data.copy()
df 
```

```python
purchase_prices = df["purchase_price"] 

avg_price = np.average(purchase_prices)
med_price = np.median(purchase_prices) 
max_price = purchase_prices.max()
min_price = purchase_prices.min()  

plt.hist(purchase_prices, range=(0, 100000), bins=6,  edgecolor='black')
plt.title("Vehicle Purchase Prices")
plt.xlabel("Purchase Price Bins")
plt.ylabel("Purchase Price Count") 

plt.axvline(avg_price, color='r', linestyle='solid', linewidth=3, label="Mean")
plt.axvline(med_price, color='y', linestyle='dotted', linewidth=3, label="Median")
plt.legend()

plt.show() 
```

```python
customer_ages = df["customer_age"] 

avg_age = np.average(customer_ages)
med_age = np.median(customer_ages) 
max_age = customer_ages.max()
min_age = customer_ages.min()  

plt.hist(customer_ages, range=(0, 100), bins=10,  edgecolor='black')
plt.title("Customer Ages")
plt.xlabel("Customer Age Bins")
plt.ylabel("Customer Age Count") 

plt.axvline(avg_age, color='r', linestyle='solid', linewidth=3, label="Mean")
plt.axvline(med_age, color='y', linestyle='dotted', linewidth=3, label="Median")
plt.legend()

plt.show() 
```

```python
customer_incomes = df["customer_income"] 

avg_income = np.average(customer_incomes)
med_income = np.median(customer_incomes)  

plt.hist(customer_incomes, range=(0, 200000), bins=10,  edgecolor='black')
plt.title("Customer Incomes")
plt.xlabel("Customer Income Bins")
plt.ylabel("Customer Income Count") 

plt.axvline(avg_income, color='r', linestyle='solid', linewidth=3, label="Mean")
plt.axvline(med_income, color='y', linestyle='dotted', linewidth=3, label="Median")
plt.legend()

plt.show() 
```
